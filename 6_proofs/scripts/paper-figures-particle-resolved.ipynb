```python
def computeAeroParticleCriticalDiameter(Dp, kappa):
    CRIT_DIAM_MAX_ITER = 100    
    sigma_sa = 0.072 # J m^-2 (Petters and Kreidenweis)
    T = 298.15 # K
    M_w = 18.01 # g mol^-1
    R = 8.3145 # J mol^-1 K^-1
    rho_w = 1e6 # g m^-3

    A = (4*sigma_sa*M_w)/(R*T*rho_w)
    dry_diam = Dp 

    if kappa < 1e-30:
        aero_particle_crit_diam = dry_diam
        return aero_particle_crit_diam
    
    c4 = -3*dry_diam**3*kappa / A
    c3 = -1*dry_diam**3*(2-kappa)
    c0 = dry_diam**6*(1-kappa)
    
    # Newtons method for f(d) = d^6 + c4 d^4 + c3 d^3 + c0
    d = max(np.sqrt((-4/3)*c4), (-1*c3**(1/3)))
    for i in range(CRIT_DIAM_MAX_ITER):
        f = d**6 + c4*d**4 + c3*d**3 + c0
        df = 6*d**5 + 4*c4*d**3 + 3*c3*d**2
        dd = f / df
        d = d - dd
        if np.abs(dd/d) < 1e-14:
            break

    return d

def computeAeroParticleCriticalRelHumid(Dp, kappa):
    sigma_sa = 0.072 # J m^-2 (Petters and Kreidenweis)
    T = 293.5 # K  (approximate temperature at inversion layer at t = 6 hr)
    M_w = 18.01 # g mol^-1
    R = 8.3145 # J mol^-1 K^-1
    rho_w = 1e6 # g m^-3

    A = (4*sigma_sa*M_w)/(R*T*rho_w)
    dry_diam = Dp 
    crit_diam = computeAeroParticleCriticalDiameter(Dp, kappa)

    if kappa < 1e-30:
        aero_particle_crit_rel_humid = np.exp(A/crit_diam)
        return aero_particle_crit_rel_humid
    else:
        aero_particle_crit_rel_humid = (crit_diam**3 - dry_diam**3) / (crit_diam**3 - dry_diam**3 * (1 - kappa))*np.exp(A/crit_diam)
        return aero_particle_crit_rel_humid

def plotThreePanel2DKappaDist(xstart, xend, ystart, yend, z_idx, **kwargs):

    fontsize = kwargs.get('global_fontsize', 9)
    plt.rcParams.update({'font.size': fontsize})

    fig = plt.figure(figsize=(6.5, 4.5))
    gs = fig.add_gridspec(2, 2, hspace=0.3, wspace=0.3,
                          left=0.12, right=0.85, bottom=0.12, top=0.92)

    ax0 = fig.add_subplot(gs[0, 0])   # top left
    ax1 = fig.add_subplot(gs[0, 1])   # top right
    ax2 = fig.add_subplot(gs[1, 1])   # bottom right
    axs_list = [ax0, ax1, ax2]

    fig.canvas.draw()

    top_pos = ax1.get_position()
    bot_pos = ax2.get_position()
    center_y = (top_pos.y1 + bot_pos.y0) / 2
    ax0_pos = ax0.get_position()

    ax0.set_position([ax0_pos.x0, center_y - ax0_pos.height / 2,
                      ax0_pos.width, ax0_pos.height])

    for j, ax in enumerate(axs_list):
        if j == 0:
            s = 'no-heterogeneity'
            t = 0
        else:
            t = 36
            if j == 1:
                s = 'no-heterogeneity'
            else:
                s = 'high-heterogeneity'

        GriddedOutput.loadData(s, xstart, xend, ystart, yend, z_idx, t)

        kappa = GriddedOutput.gridded_data[s]['aero_kappa']
        diams = GriddedOutput.gridded_data[s]['aero_diams']
        numconc = GriddedOutput.gridded_data[s]['aero_numconc']
        n_cells = GriddedOutput.gridded_data[s]['n_total_cells']

        diam_bins = np.logspace(-8, -6, 50)
        kappa_bins = np.linspace(0, 1, 51)
        diam_logwidth = (np.log10(diam_bins[1:]) - np.log10(diam_bins[0:-1]))[0]
        kappa_width = (kappa_bins[1:] - kappa_bins[0:-1])[0]
        hist, xedges, yedges, cs = ax.hist2d(diams, kappa, bins=[diam_bins, kappa_bins],
                                             weights=numconc / (diam_logwidth * n_cells),
                                             density=False, cmin=1, edgecolor='face', linewidth=.1,
                                             norm=mplcolors.LogNorm(1e7, 1e10), zorder=2)
        ax.set_xscale('log')

        if kwargs.get('xlims'):
            ax.set_xlim(kwargs.get('xlims'))

        # X-axis labels: hide on top-right panel
        if j == 1:
            ax.set_xticklabels([])
        else:
            ax.set_xlabel(r'$D_{\mathrm{p}}$ (m)')

        # Y-axis labels: only on left column and bottom-right
        if j in [1, 2]:
            pass
        else:
            ax.set_ylabel(r'$\kappa$')

        # Panel titles
        if j == 0:
            text = 'Initial condition'
        elif j == 1:
            text = 'No heterogeneity, $t=6$ h'
        else:
            text = 'High heterogeneity, $t=6$ h'

        txt = AnchoredText(f'{text}',
                           loc='lower left', prop=dict(size=fontsize), frameon=True,
                           bbox_to_anchor=(0., 1.), borderpad=0, pad=0.25,
                           bbox_transform=ax.transAxes)
        ax.add_artist(txt)

        # Grid
        ax.grid(which="major", linewidth=1, axis='y', ls="dotted", dashes=(.4, 6), c='#414141', alpha=.5, zorder=0)
        ax.grid(which="major", linewidth=1, axis='x', ls="dotted", dashes=(.4, 6), c='#414141', zorder=0)
        ax.grid(which="minor", linewidth=.8, axis='y', ls="dotted", dashes=(.4, 4), c='white', zorder=0)
        ax.grid(which="minor", linewidth=.8, axis='x', ls="dotted", dashes=(.4, 4), c='#414141', zorder=0)

        # Arrows from initial condition to scenario plots
        if j == 0:
            fig.canvas.draw()
            ax0_pos = ax0.get_position()
            ax1_pos = ax1.get_position()
            ax2_pos = ax2.get_position()

            arrow_up = mpatches.FancyArrowPatch(
                (ax0_pos.x1-0.05, ax0_pos.y1+0.05),
                (ax1_pos.x0-0.07, ax1_pos.y0 + 0.66*ax1_pos.height),
                lw=.1, facecolor='k',
                arrowstyle='Simple,head_width=10,head_length=10,tail_width=2',
                transform=fig.transFigure)
            arrow_up.set_clip_on(False)
            fig.patches.append(arrow_up)

            arrow_down = mpatches.FancyArrowPatch(
                (ax0_pos.x1-0.05, ax0_pos.y0-0.05),
                (ax2_pos.x0-0.07, ax2_pos.y0 + (1-0.66)*ax2_pos.height),
                lw=.1, facecolor='k',
                arrowstyle='Simple,head_width=10,head_length=10,tail_width=2',
                transform=fig.transFigure)
            arrow_down.set_clip_on(False)
            fig.patches.append(arrow_down)

        # Supersaturation contours
        contour_diams = np.logspace(-8, -6, 300)
        contour_kappas = np.linspace(0, 1, 300)

        Sc_map = np.zeros((len(contour_diams), len(contour_kappas)))
        for ii, diam in enumerate(contour_diams):
            for jj, kap in enumerate(contour_kappas):
                Sc_map[ii, jj] = computeAeroParticleCriticalRelHumid(diam, kap)

        contour_levels = [1.001, 1.003, 1.006, 1.01]
        contours = ax.contour(contour_diams, contour_kappas, Sc_map.T,
                              vmin=1, vmax=1.01, linewidths=1.5,
                              levels=contour_levels, colors=['k'], zorder=3)

        fmt = {l: f'{100 * (l - 1):3.1f}%' for l in contour_levels}
        ax.clabel(contours, inline=True, fontsize=fontsize, fmt=fmt,
                  manual=[(2.5e-8, 0.76), (3.5e-8, 0.78), (7e-8, 0.82), (1.5e-7, 0.86)])

    # Colorbar: manually positioned to span the right column
    fig.canvas.draw()
    right_top = ax1.get_position()
    right_bot = ax2.get_position()
    cbar_ax = fig.add_axes([right_top.x1 + 0.03, right_bot.y0, 0.025, right_top.y1 - right_bot.y0])
    cbar = fig.colorbar(cs, cax=cbar_ax, orientation='vertical')
    cbar.set_label(label='Number conc. (m$^{-3}$)')

    if kwargs.get("savefig"):
        plt.savefig(f'2d-kappa-dist-three-panel-z{z_idx}.pdf', format='pdf', bbox_inches='tight')
```
